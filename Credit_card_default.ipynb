# Data Preprocessing and Cleaning

```python
df.head()
```

```python
df['EDUCATION'] = df['EDUCATION'].replace([0,4,5,6] , 4 , inplace=True)
```

```python
df.drop(columns=['ID' , 'EDUCATION'] , inplace=True)
```

```python
df['MARRIAGE'] = df['MARRIAGE'].replace(0, 3)
```

```python
df['MARRIAGE'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = df.drop(columns=['default.payment.next.month'] , axis=1)
y = df['default.payment.next.month']
```

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
models = {
    'Logistic Regression' : LogisticRegression(),
     'GaussianNB' : GaussianNB(),
     'Knn' : KNeighborsClassifier(),
     'DecisionTreeClassifier' : DecisionTreeClassifier(),
      'SVM' : SVC()
}
```

```python
for name, model in models.items(): 
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name} is {acc:.5f}") 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l2'],
    'solver': ['liblinear', 'lbfgs']  # both work with l2
}
```

```python
grid = GridSearchCV(
    LogisticRegression(),
    param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid.fit(X_train, y_train)
```

```python
pd.DataFrame(grid.cv_results_)
```
